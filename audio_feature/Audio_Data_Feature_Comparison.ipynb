```python
import librosa
from librosa import display as dp
import matplotlib.pyplot as plt
from veiw_audio import normalize
from extract_feature import ExtractFeature
```

```python
audio, sr = librosa.load(path='./dataset/train/bed/00f0204f_nohash_0.wav', sr=None)
ef = ExtractFeature(audio=audio, sample_rate=sr, n_fft=256)
ef.display_feature()
```

```python
normalized_audio = normalize(audio)
ef_norm = ExtractFeature(audio=normalized_audio, sample_rate=sr, n_fft=256)
ef_norm.display_feature()
```

```python
audio, sr = librosa.load(path='./dataset/train/bed/00f0204f_nohash_0.wav', sr=None)
ef = ExtractFeature(audio=audio, sample_rate=sr, n_fft=512)
ef.display_feature()
```

```python
normalized_audio = normalize(audio)
ef_norm = ExtractFeature(audio=normalized_audio, sample_rate=sr, n_fft=512)
ef_norm.display_feature()
```

# 결론
* fft length가 256일때는 Spectrogram에만 변화가 보이다가 512로 늘어나자 Mel-Spectrogram에서도 비슷한 거동을 보인다.
* 나머지 특징 값들은 정규화와 fft length에 대해서 큰 변화가 관찰되지 않는다.

이것은 정규화 과정에서 전체 신호의 크기를 키웠기 때문이다. 그 결과로 신호의 진폭의 범위가 상대적으로 넓어지면서 얻어진 결과물이다.
